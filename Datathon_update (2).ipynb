#### some extra EDA

```python
# --- Analisis 2: Analisis Jumlah Komorbiditas ---
print("\nAnalisis 2: Pengaruh Jumlah Komorbiditas terhadap Length of Stay")
# Daftar kolom kondisi sesuai yang ada di dokumen
condition_cols = [
    'dialysisrenalendstage', 'asthma', 'irondef', 'pneum', 'substancedependence',
    'psychologicaldisordermajor', 'depress', 'psychother', 'fibrosisandother',
    'malnutrition', 'hemo'
]

data['num_conditions'] = data[condition_cols].sum(axis=1)

plt.figure(figsize=(14, 8))
sns.boxplot(x='num_conditions', y='lengthofstay', data=data, palette='plasma')
plt.title('Hubungan Antara Jumlah Komorbiditas dan Lama Menginap', fontsize=16)
plt.xlabel('Jumlah Total Kondisi Medis', fontsize=12)
plt.ylabel('Lama Menginap (Hari)', fontsize=12)
plt.savefig('komorbid vs LOS.png', bbox_inches='tight')
plt.show()
```

```python
# --- Analisis 4: Tren Waktu (Time Series) Jumlah Pasien Masuk ---
print("\nAnalisis 4: Tren Jumlah Pasien Masuk per Bulan")
data['month_year'] = data['vdate'].dt.to_period('M')
admissions_by_month = data['month_year'].value_counts().sort_index()

plt.figure(figsize=(15, 7))
admissions_by_month.plot(kind='line', marker='o', linestyle='-')
plt.title('Jumlah Pasien Masuk per Bulan', fontsize=16)
plt.xlabel('Bulan-Tahun', fontsize=12)
plt.ylabel('Jumlah Pasien', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.savefig('jumlah pasien masuk perbulan.png', bbox_inches='tight')
plt.show()
```

```python
# --- Analisis 6: Visualisasi Hubungan Antar Variabel Klinis dengan Pair Plot ---
print("Analisis 6: Membuat Pair Plot untuk melihat hubungan antar variabel klinis utama.")

# Membuat kategori 'lengthofstay' untuk visualisasi yang lebih jelas
bins = [0, 3, 7, data['lengthofstay'].max()]
labels = ['Pendek (0-3 hari)', 'Menengah (4-7 hari)', 'Panjang (>7 hari)']
data['los_category'] = pd.cut(data['lengthofstay'], bins=bins, labels=labels, right=True)

# Pilih beberapa variabel klinis kunci untuk dianalisis
pairplot_vars = ['hematocrit', 'glucose', 'bmi', 'pulse', 'los_category']

# Buat pairplot
sns.pairplot(data[pairplot_vars], hue='los_category', palette='viridis', corner=True)
plt.suptitle('Hubungan Antar Variabel Klinis Berdasarkan Kategori Lama Rawat Inap', y=1.02, fontsize=16)
plt.savefig('pairplot_klinis.png', bbox_inches='tight')
plt.show()
```

```python
# --- Analisis 7: Pengaruh Riwayat Kunjungan Ulang (rcount) ---
print("\nAnalisis 7: Menganalisis pengaruh riwayat kunjungan ulang (rcount) terhadap lama rawat inap.")

plt.figure(figsize=(12, 8))
sns.boxplot(data=data, x='rcount', y='lengthofstay', palette='coolwarm')
plt.title('Distribusi Lama Rawat Inap Berdasarkan Riwayat Kunjungan Ulang (rcount)', fontsize=16)
plt.xlabel('Jumlah Kunjungan Ulang Sebelumnya', fontsize=12)
plt.ylabel('Lama Rawat Inap (Hari)', fontsize=12)
plt.savefig('length vs rcount.png', bbox_inches='tight')
plt.show()
```

#### Split Dataset


```python
X = data.select_dtypes(include='number').drop(columns=['lengthofstay'])
y = data['lengthofstay']
```

```python
from sklearn.model_selection import train_test_split

# Langkah 1: Bagi data menjadi train dan test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=4)

# Langkah 2: Bagi data training menjadi train dan validation
X_train, X_val, y_train, y_val = train_test_split(
    X_train, y_train, test_size=0.25, random_state=4
)
```

#### Polynomial Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

# Buat model polynomial derajat 2 (misal)
degree = 2
model = make_pipeline(PolynomialFeatures(degree), LinearRegression())

# Fit ke data latih
model.fit(X_train, y_train)

# Prediksi di data validasi
y_pred = model.predict(X_val)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np


mse = mean_squared_error(y_val, y_pred)
r2 = r2_score(y_val, y_pred)

print(f"Mean Squared Error: {mse:.2f}")
rmse = np.sqrt(mse)
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"R² Score: {r2:.2f}")
```

#### Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
import matplotlib.pyplot as plt

# 1. Buat dan latih model Linear Regression
model = LinearRegression()
model.fit(X_train, y_train)

# 2. Prediksi di data validasi
y_pred = model.predict(X_val)

# 3. Evaluasi: MSE, RMSE, R²
mse = mean_squared_error(y_val, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_val, y_pred)

print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"R² Score: {r2:.2f}")
```

#### Random Forest Model

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Buat model Random Forest
model = RandomForestRegressor(n_estimators=100, random_state=42)

# Latih model
model.fit(X_train, y_train)

# Prediksi pada data validasi
y_pred = model.predict(X_val)

# Evaluasi performa
mse = mean_squared_error(y_val, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_val, y_pred)

print(f"MSE  : {mse:.2f}")
print(f"RMSE : {rmse:.2f}")
print(f"R²   : {r2:.2f}")
```

#### SVR

```python
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
import numpy as np

# ⚠️ Penting: SVR butuh fitur yang diskalakan (scaling!)
scaler_X = StandardScaler()
scaler_y = StandardScaler()

# Scaling fitur dan target
X_train_scaled = scaler_X.fit_transform(X_train)
X_val_scaled = scaler_X.transform(X_val)

y_train_scaled = scaler_y.fit_transform(y_train.values.reshape(-1, 1)).ravel()

# Buat dan latih model SVR (pakai kernel RBF default)
model = SVR(kernel='rbf', C=1.0, epsilon=0.1)
model.fit(X_train_scaled, y_train_scaled)

# Prediksi dan kembalikan ke skala awal
y_pred_scaled = model.predict(X_val_scaled)
y_pred = scaler_y.inverse_transform(y_pred_scaled.reshape(-1, 1)).ravel()
```

```python
mse = mean_squared_error(y_val, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_val, y_pred)

print(f"MSE  : {mse:.2f}")
print(f"RMSE : {rmse:.2f}")
print(f"R²   : {r2:.2f}")
```

#### Regularisasi Model LightGBM

```python
from sklearn.model_selection import GridSearchCV
from lightgbm import LGBMRegressor

# Buat parameter grid
param_grid = {
    'n_estimators': [100, 200],
    'learning_rate': [0.01, 0.05, 0.1],
    'max_depth': [3, 5, 7],
    'min_child_weight': [1, 5, 10],
    'lambda_l1': [0, 0.1, 1.0],
    'lambda_l2': [0, 0.1, 1.0],
    'min_split_gain': [0.0, 0.01, 0.1]
}

# Inisialisasi model
model = LGBMRegressor()

# Setup GridSearchCV
grid_search = GridSearchCV(
    estimator=model,
    param_grid=param_grid,
    cv=5,                      # 5-fold cross-validation
    scoring='r2',              # Bisa juga pakai 'neg_mean_squared_error'
    verbose=1,
    n_jobs=-1                  # Gunakan semua core CPU
)

# Jalankan tuning
grid_search.fit(X_train, y_train)

# Cetak hasil terbaik
print("Best parameters:", grid_search.best_params_)
print("Best R² score (CV):", grid_search.best_score_)

# Evaluasi di data validasi
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_val)

mse = mean_squared_error(y_val, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_val, y_pred)

print(f"\n[Evaluasi Model Terbaik di Validasi]")
print(f"MSE  : {mse:.2f}")
print(f"RMSE : {rmse:.2f}")
print(f"R²   : {r2:.2f}")
```

#### Evaluasi Data Test

```python
# Evaluasi di data test
y_test_pred = best_model.predict(X_test)

mse_test = mean_squared_error(y_test, y_test_pred)
rmse_test = np.sqrt(mse_test)
r2_test = r2_score(y_test, y_test_pred)

print(f"\n[Evaluasi Model Terbaik di Data Test]")
print(f"MSE  : {mse_test:.2f}")
print(f"RMSE : {rmse_test:.2f}")
print(f"R²   : {r2_test:.2f}")
```

```python
from sklearn.model_selection import learning_curve
import matplotlib.pyplot as plt
import numpy as np

# Gunakan model terbaik dari GridSearch
model = best_model

# Hitung learning curve
train_sizes, train_scores, val_scores = learning_curve(
    model, X_train, y_train,
    cv=5,
    scoring='r2',
    train_sizes=np.linspace(0.1, 1.0, 10),
    n_jobs=-1
)

# Rata-rata dan standar deviasi
train_scores_mean = np.mean(train_scores, axis=1)
train_scores_std  = np.std(train_scores, axis=1)
val_scores_mean   = np.mean(val_scores, axis=1)
val_scores_std    = np.std(val_scores, axis=1)

# Plot learning curve
plt.figure(figsize=(8, 6))
plt.plot(train_sizes, train_scores_mean, 'o-', color='blue', label='Training score')
plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                 train_scores_mean + train_scores_std, alpha=0.1, color='blue')

plt.plot(train_sizes, val_scores_mean, 'o-', color='green', label='Validation score')
plt.fill_between(train_sizes, val_scores_mean - val_scores_std,
                 val_scores_mean + val_scores_std, alpha=0.1, color='green')

plt.title('Learning Curve (LightGBM)')
plt.xlabel('Training Set Size')
plt.ylabel('R² Score')
plt.legend(loc='best')
plt.grid(True)
plt.tight_layout()
plt.show()
```
